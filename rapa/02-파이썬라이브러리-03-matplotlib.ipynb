```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

### 선그래프

```python
a = np.array([10,14,19,20,25])
print(a)
plt.plot(a)

a2 = np.arange(20)
print(a2)
plt.plot(a2)

plt.show() # 새 프레임에 그래프를 출력
```

```python
# 1차 방정식의 그래프 : f(x) = a*x + b , 직선의 방정식
x = np.arange(-4.5,5,0.5)
y = 2*x + 0
print(x,y)
plt.plot(x,y)
plt.show()
```

```python
# 2차 방정식의 그래프 : f(x) = a*x^2 + b, 포물선의 방정식
x = np.arange(-4.5,5,0.5)
y = 2*x**2
print(x,y)
plt.plot(x,y)

plt.plot(x,4*x)
plt.plot(x,-3*x)
plt.show()
```

### 로그플롯(log plot)

```python
print(np.log(100)) # 자연로그,  e: 2.71828
print(np.exp(4.605170185988092))  # 100
# log(100) = 4.605170185988092
# 100 =  e**4.605170185988092
```

```python
df = pd.read_csv('transcount.csv')
df = df.groupby('year').agg(np.mean)
df
```

```python
years = df.index.values 
counts = df['trans_count'].values
print(type(years),type(counts))
print(years,counts)
plt.plot(years,counts) # 원본 데이터로 그래프를 출력 , 
                       # 올바른 정보(변화량)를 알 수가 없다
                       # 1970~2000년 사이의 변화량이 매우 작음
plt.show()
```

```python
# y 값을 log변환하여 그래프로 출력
plt.plot(years,np.log(counts)) # 자연로그, e:2.71828
                               # 1970~2000년 사이의 변화량이 분명해짐
plt.show()
```

```python
print(np.log1p(0))  # 1을 더한 후 log변환 ,log(0)이 -inf(-무한대)가 되지 않도록
print(np.log(1))    
# log(0) => -inf
# log(1) => 0

# np.log1p(x) = np.log(0 + 1)
```

```python
### 로그 플롯
years = df.index.values
counts = df['trans_count'].values

plt.semilogy(years,counts,'o')  # semilogy함수는 y 값을 로그변환하여 그래프로 출력

# np.polyfit(x,y,...):x,y값을 사용하여 오차가 가장 적은 직선의 기울기와 y절편을구한다
#                     예측 방정식(H(x) = w*x + b)을 얻는다
poly = np.polyfit(years,np.log(counts),deg=1) # deg=1: x는 1차원
print(poly)  # [ 3.61559210e-01 -7.05783195e+02] , 회귀 계수
# H(x) = 3.61559210e-01*x -7.05783195e+02 ,예측 방정식

plt.semilogy(years,np.exp(np.polyval(poly,years))) # 노란색 직선 ,예측 함수의 그래프

plt.show()

np.exp(np.polyval(poly,years)) # polyfit()으로 얻은 poly내의 기울기와 y절편을 이용하여
# x값을 넣어 예측값(H(x), y축의 값)을 구한다
```

```python
## 지연(Lag) 플롯 : time series data(시계열 데이터) 분석

from pandas.plotting import lag_plot

# 지연값 : 1년 
# x축 : 기준년도의  trans_count 로그값
# y축 : 기준년도 + 1의 trans_count 로그값
lag_plot(np.log(df['trans_count']))
plt.grid()
plt.show()

df
```
