```python
inStr = '''코스모스는 과거에 있었고, 현재에 있으며, 미래에 있을 그 모든 것이다.
우리는 코스모스의 일부다.
이것은 결코 시적 수사가 아니다.'''
# 방법1:
def  letterCount():
    #global 안해도 함수안에서 쓸 수 있긴 함.
    global inStr
    countDic = {}
    countList=[]  #없어도 상관없을듯
    # 글자:빈도수 딕셔너리 만들기
    for char in inStr:
        if char >= '가' and char <='힣':
            #문자열의 해당 문자가 한글인데 기존 딕셔너리 안에 있으면 카운팅
            if char in countDic:
                countDic[char] += 1
            #문자열의 해당 문자가 한글인데 기존 딕셔너리 안에 없으면 새로 키 생성.
            else:
                countDic[char] = 1
    # 정렬하기
    # 방법 1 . reverse= False(오름차순)가 기본값임. x[0]은 키, x[1]은 밸류. 밸류로 정렬함.
    #seorted 함수에서 key=값은 정렬기준이 되는 변수. 근데 x[1]이 함수인가? 
    countList = sorted(countDic.items(), key=lambda x: x[1], reverse=True)
    # 방법 2
    import operator
    countList = sorted(countDic.items(), key=operator.itemgetter(1), reverse=True)
    return countList
print(letterCount())
```

```python
#온라인 장바구니 구현

class Basket : 
   
    def __init__(self,id):
        self.id = id
        self.items = []
        self.prices = []
        self.qtys = []
        self.totprice = 0
        self.itemcnt = 0
        
    #장바구니에 추가
    def add(self, item, price, qty):
        self.items.append(item)
        self.prices.append(price)
        self.qtys.append(qty)
        self.totprice += price*qty
        self.itemcnt += 1
        
    #장바구니에서 삭제
    #리스트에서는 수량이 0 이하가 되었을 때 그 항목을 삭제하기가 복잡하다. 
    def delete(self, item, qty):
        for i in range (len(self.items)):
            # 0이면 삭제하기.
            if self.qtys[i]=<qty:
                self.itemcnt -= 1
                del(self.items[i])
                del(self.prices[i])
                del(self.qtys[i])
            elif item == self.items[i] :
                self.totprice -= self.prices[i]*qty
                self.qtys[i] -= qty
  

                
                
        
    #장바구니에 물건 출력
    def printitems(self):
        print(f"#{self.id}의 장바구니")
        print("-"*20)
        for i , v in enumerate(self.items):
            print(f'*  {v:5} {self.prices[i]:5} {self.qtys[i]:2}')
        print(f'** 전체가격 = {v} {self.totprice:5} {self.itemcnt:2}')
        print("-"*20)
        

        
basket1 = Basket("제인")
basket2 = Basket("대원")

basket1.add("바나나", 3000, 2)
basket1.add("우유", 2500, 2)
basket1.add("라면", 4000, 1)
basket2.add("라면", 4000, 2)
basket1.delete("우유",3)

basket1.printitems()
basket2.printitems()
```

```python
class Basket:
    def __init__(self, id):
        self.id = id
        self.바구니={}
    def add(self, item, price, qty):
        self.바구니[item]=[price,qty]
    def delete(self, item, qty):
        if self.바구니[item][1] >= qty :
            self.바구니[item][1] -= qty
        else : 
            del(self.바구니[item])
            
    def printitems(self):
        print(f'# {self.id}의 장바구니')
        print('-'*30)
        total = 0
        cnt = len(self.바구니)
        for key in self.바구니:
            print(f'* {key:10} {self.바구니[key][0]} {self.바구니[key][1]}')
            total += self.바구니[key][0] * self.바구니[key][1]
        print(f'** 전체가격 =  {total} 전체 수량 = {cnt}')
basket1 = Basket("제인")
basket2 = Basket("대원")
basket1.add('바나나',3000,2)
basket1.add('우유',2500,2)
basket1.add('라면',4000,1)
basket2.add('라면',4000, 2)
basket1.delete('우유',3)
basket1.printitems()
basket2.printitems()
```
